```python
import csv
import traceback
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

try:
    # Initialize Chrome WebDriver
    options = webdriver.ChromeOptions()
    options.add_argument('--disable-blink-features=AutomationControlled')
    driver = webdriver.Chrome(options=options)

    #  Source URL 
    driver.get("https://www.imdb.com/chart/top/") 

    # Wait for the list to appear
    WebDriverWait(driver, 40).until(EC.visibility_of_element_located((By.CSS_SELECTOR, ".compact-list-view")))
    
    # Fetch all rows
    rows = driver.find_elements(By.CSS_SELECTOR, '.compact-list-view li')
    
    # Open a CSV file in write mode
    with open('imdb_top_movies.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        
        # Write header row
        writer.writerow(['Title', 'Year', 'Duration', 'Rating'])
        
        # Iterate through rows
        for row in rows:
            # Extract title, year, duration, and rating from each record
            title = row.find_element(By.CSS_SELECTOR, '.ipc-title a').text
            year = row.find_element(By.CSS_SELECTOR, '.cli-title-metadata span:nth-child(1)').text
            duration = row.find_element(By.CSS_SELECTOR, '.cli-title-metadata span:nth-child(2)').text
            rating = row.find_element(By.CSS_SELECTOR, '.ratingGroup--imdb-rating').text
            print("title=>",title,"year=>",year,"duration=>",duration,'rating=>',rating)
            
            # Write data to CSV file
            writer.writerow([title, year, duration, rating])
            
    print("Data saved to imdb_top_movies.csv")
    
except Exception as e:
    traceback.print_exc()

finally:
    # Close the WebDriver session
    driver.quit()
```

```python
import traceback
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

try:
    # Initialize Chrome WebDriver
    options = webdriver.ChromeOptions()
    options.add_argument('--disable-blink-features=AutomationControlled')
    driver = webdriver.Chrome(options=options)

    # Source URL 
    driver.get("https://space-facts.com/mars/") 

    # Wait for the list to appear
    WebDriverWait(driver, 40).until(EC.visibility_of_element_located((By.CSS_SELECTOR, "#tablepress-p-mars tbody tr td:nth-child(2)")))
    
    # Fetch all rows
    data = {
        "Attribute": ["Equatorial Diameter", "Polar Diameter", "Mass", "Moons", "Orbit Distance", 
                      "Orbit Period", "Surface Temperature", "First Record", "Recorded By"],
        "Value": [
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(1) td:nth-child(2)').text,
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(2) td:nth-child(2)').text,
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(3) td:nth-child(2)').text,
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(4) td:nth-child(2)').text,
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(5) td:nth-child(2)').text,
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(6) td:nth-child(2)').text,
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(7) td:nth-child(2)').text,
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(8) td:nth-child(2)').text,
            driver.find_element(By.CSS_SELECTOR, '#tablepress-p-mars tbody tr:nth-child(9) td:nth-child(2)').text
        ]
    }

    # Create DataFrame
    df = pd.DataFrame(data)

    # Save DataFrame to Excel
    df.to_excel("mars_data.xlsx", index=False)

except Exception as e:
    traceback.print_exc()

finally:
    # Close the WebDriver session
    driver.quit()
```
